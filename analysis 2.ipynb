```python
time_dic = {}
for i in access_dict:
    time_dic[i] = {'first': [], 'second': [], 'last': []}
    time_dic[i]['first'] = access_dict[i]['times_1']
    try:
        time_dic[i]['second'] = access_dict[i]['times_2']
    except:
         time_dic[i]['second'] = access_dict[i]['times_1']
    print(list(access_dict[i].keys())[-1])
    time_dict[i]['last'] = access_dict[i][list(access_dict[i].keys())[-1]]
    
```

```python
from sklearn.metrics import mutual_info_score

def calc_MI(x, y, bins):
    c_xy = np.histogram2d(x, y, bins)[0]
    mi = mutual_info_score(None, None, contingency=c_xy)
    return mi
```

```python
mi = calc_MI(access_dict['Paris']['times_1'],access_dict['Paris']['times_2'],20)
print(mi)
```

```python
mi_list_12 = []
mi_list_11 = []
city_list = []
for city in time_dict:
    x = time_dict[city]['first']
    y = time_dict[city]['second']
    z = time_dict[city]['last']
    mi_list_12.append(calc_MI(x, y, 20))
    mi_list_11.append(calc_MI(x, z, 20))
    city_list.append(city)
   
```

```python
import matplotlib.pyplot as plt
#plt.scatter(mi_list_12,mi_list_11)
fig, ax = plt.subplots()
ax.scatter(mi_list_12,mi_list_11)
plt.xlabel("MI between first and second closseset centers")
plt.ylabel("MI between first and last closseset centers")
for txt in city_list:
        ax.annotate(txt, (mi_list_12[city_list.index(txt)], mi_list_11[city_list.index(txt)]))
plt.show()
```
