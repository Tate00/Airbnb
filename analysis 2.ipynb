```python
time_dict = {}
for i in access_dict:
    time_dict[i] = {'first': [], 'second': [], 'last': []}
    time_dict[i]['first'] = access_dict[i]['times_1']
    try:
        time_dict[i]['second'] = access_dict[i]['times_2']
    except:
         time_dict[i]['second'] = access_dict[i]['times_1']
    #print(list(access_dict[i].keys())[-1])
    time_dict[i]['last'] = access_dict[i][list(access_dict[i].keys())[-1]]
    
```

```python
from sklearn.metrics import mutual_info_score

def calc_MI(x, y, bins):
    c_xy = np.histogram2d(x, y, bins)[0]
    mi = mutual_info_score(None, None, contingency=c_xy)
    return mi
```

```python
mi = calc_MI(access_dict['Paris']['times_1'],access_dict['Paris']['times_2'],20)
mi1 = calc_MI(access_dict['Paris']['times_1'],access_dict['Paris']['times_3'],20)
print(mi)
print(mi1)
```

```python
mi_list_12 = []
mi_list_11 = []
city_list = []
for city in time_dict:
    x = time_dict[city]['first']
    y = time_dict[city]['second']
    z = time_dict[city]['last']
    mi_list_12.append(calc_MI(x, y, 20))
    mi_list_11.append(calc_MI(x, z, 20))
    city_list.append(city)
print(mi_list_12)
print(mi_list_11)
print(city_list)
```

```python
import matplotlib.pyplot as plt
#plt.scatter(mi_list_12,mi_list_11)
fig, ax = plt.subplots()
ax.scatter(mi_list_12,mi_list_11)
plt.xlabel("MI between first and second closseset centers")
plt.ylabel("MI between first and last closseset centers")
for txt in city_list:
        ax.annotate(txt, (mi_list_12[city_list.index(txt)], mi_list_11[city_list.index(txt)]))
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
X = np.column_stack((mi_list_11, eta_list))
X
```

```python
kmeans_kwargs = {
            "init": "random",
            "n_init": 10,
            "max_iter": 300,
            "random_state": 42,
        }
       
# A list holds the SSE values for each k
sse = []
for k in range(1, 15):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 15), sse)
plt.xticks(range(1, 15))
plt.xlabel("Number of clusters")
plt.ylabel("sse")
plt.show()
```

```python
kl = KneeLocator(
    range(1, 15), sse, curve="convex", direction="decreasing")

kl.elbow
```

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    
    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)
        
        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels

centers, labels = find_clusters(X, 3)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

```python
import matplotlib.pyplot as plt
#plt.scatter(mi_list_12,mi_list_11)
fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c=labels,s=50, cmap='viridis')
plt.xlabel("MI between first and last closseset centers")
plt.ylabel("Eta")
for txt in city_list:
        ax.annotate(txt, (X[:,0][city_list.index(txt)], X[:,1][city_list.index(txt)]))
plt.show()
```
